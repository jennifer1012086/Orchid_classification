```python
IMG_SIZE = 150
model = keras.models.load_model('./trained_model.h5')
```

```python
fns_all = os.listdir("./orchid_private_set/")
```

```python
def make_train_data(DIR, files):
    x_test = []
    fns = files

    for img in files:
        path = os.path.join(DIR,img)
        img = cv2.imread(path,cv2.IMREAD_COLOR)
        img = cv2.resize(img, (IMG_SIZE,IMG_SIZE))
        
        x_test.append(np.array(img))

    return np.array(x_test), fns
```

```python
### public set
x, fns = make_train_data("./orchid_private_set", fns_all[22500:23750])
x = x / 255
pred = model.predict(x)
pred_digits = np.argmax(pred,axis=1)

data = {"filename": fns, "category": list( map(int, pred_digits) )}
dfPb = pd.DataFrame(data)
dfPb.to_csv("./private_22500_23750_yt.csv", index=False)
# dfPb.to_csv("./public.csv", mode='a', index=False, header=False)
```
